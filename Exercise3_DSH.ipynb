```python
# 2. Reading example accelaration signal and labels
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
acc = pd.read_csv('acc_exp01_user01.txt', header = None, sep = ' ')
labels = pd.read_csv('labels.txt', header = None, sep = ' ')
x_acc = acc.loc[:,0]
y_acc = acc.loc[:,1]
z_acc = acc.loc[:,2]
starting_marker = labels.loc[:,3]
ending_marker = labels.loc[:,4]
# time = (ending_marker - starting_marker) * 0.02  Don't know how to calculate this?
fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(x_acc)
ax.plot(y_acc)
ax.plot(z_acc)
plt.xlabel('Time (samples)')
plt.ylabel('Acceleration (m/s^2)')
plt.legend(['x_acceleration','y_acceleration','z_acceleration'])
plt.show()
# In the plot we can see that x-acceleration is the greateast and all accelerations have 
# significant variation in 10000-20000 time samples. Acceleration fluctuations are due to
# for example, starting to walk or standing up.
```

```python
# 3. Calculating statistical features.
import statistics
x_mean = statistics.mean(x_acc)
y_mean = statistics.mean(y_acc)
z_mean = statistics.mean(z_acc)
print('x-acceleration mean:',x_mean)
print('y-acceleration mean:',y_mean)
print('z-acceleration mean:',z_mean)
# I chose mean, since mean gives an idea where the center value is located in dataset.

x_std = statistics.stdev(x_acc)
y_std = statistics.stdev(y_acc)
z_std = statistics.stdev(z_acc)
print('x-acceleration standard deviation:',x_std)
print('y-acceleration standard deviation:',y_std)
print('z-acceleration standard deviation:',z_std)
# Standard deviation is important, because it helps understanding when data is distributed.

x_var = statistics.variance(x_acc)
y_var = statistics.variance(y_acc)
z_var = statistics.variance(z_acc)
print('x-acceleration variance:',x_var)
print('y-acceleration variance:',y_var)
print('z-acceleration variance:',z_var)
# Variance allows to measure the dispersion of the dataset around mean.
```
